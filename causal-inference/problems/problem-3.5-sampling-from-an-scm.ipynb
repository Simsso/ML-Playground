# Problem 3.5 Sampling from an SCM

Given the SCM
\begin{align}
X:=&Y^2+N_X\\
Y:=&N_Y
\end{align}
with $N_X,N_Y\sim\mathcal{N}\left(0,1\right)$ generate an i.i.d. sample of size 200 from the joint distribution $(X,Y)$

```python
import numpy as np
```

```python
mu = 0
sigma = 1
n = 200
```

```python
n_y = np.random.normal(mu, sigma, 200)
```

```python
n_x = np.random.normal(mu, sigma, 200)
```

```python
y = n_y
```

```python
x = y**2 + n_x
```
