```python
import pandas as pd
import folium
```

```python
flight_logs = [
    'flight-data-nh211.csv',
    'flight-data-nh212.csv',
]
```

```python
# Create a map centered at an arbitrary location
m = folium.Map(location=[51.471889, -0.441647], zoom_start=3)  # Zoom out to see the entire globe

# Function to add polylines
def add_polyline(dataframe):
    locations = dataframe[['Latitude', 'Longitude']].values
    if len(locations) > 1:  # Ensure there are enough points for a line
        folium.PolyLine(locations, color="red", weight=2.5, opacity=1).add_to(m)

# Loop through each flight log
for flight_log in flight_logs:
    # Load the CSV file
    df = pd.read_csv(flight_log)
    
    # Extract coordinates from the 'Position' column and split them into latitude and longitude
    df[['Latitude', 'Longitude']] = df['Position'].str.strip('"').str.split(',', expand=True).astype(float)

    # Initialize last longitude for detection of crossing the IDL
    last_longitude = df.iloc[0]['Longitude']

    # Start of the segment
    start_idx = 0

    for idx, row in df.iterrows():
        if abs(row['Longitude'] - last_longitude) > 180:
            # Draw line for the segment before crossing the IDL
            add_polyline(df.loc[start_idx:idx-1])
            
            # Start new segment
            start_idx = idx

        # Update last longitude
        last_longitude = row['Longitude']

    # Draw line for the last segment
    add_polyline(df.loc[start_idx:])

# Display the map
m
```

```python
import json

# List of flight log files
flight_logs = [
    'flight-data-nh211.csv',
    'flight-data-nh212.csv',
]

flight_data = []

# Loop through each flight log
for flight_log in flight_logs:
    # Load the CSV file
    df = pd.read_csv(flight_log)
    
    # Extract coordinates from the 'Position' column and split them into latitude and longitude
    df[['Latitude', 'Longitude']] = df['Position'].str.strip('"').str.split(',', expand=True).astype(float)
    
    # Append the data
    flight_data.extend(df[['Latitude', 'Longitude']].values.tolist())
    
# Convert Python list of lists to JSON array of arrays
flight_data_json = json.dumps(flight_data)

print(flight_data_json)
```
